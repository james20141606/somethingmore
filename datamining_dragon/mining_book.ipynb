```python
def get_words(txt):  
    seg_list = jieba.cut(txt)  
    c = Counter()  
    for x in seg_list:  
        if len(x)>1 and x != '\r\n':  
            c[x] += 1  
    print ('常用词频度统计结果') 
    for (k,v) in c.most_common(10):  
        print('%s%s %s  %d' % ('  '*(5-len(k)), k, '*'*int(v/20), v))  
    return c
```
