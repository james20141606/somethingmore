```python
exchangerate = pd.read_excel('exchangerate.xls',index_col=0)
exrate = pd.read_excel('exrate.xls',index_col=0)
exchangerate.index = np.array([pd.to_datetime(str(i), format='%Y%m%d') for i in 
                              np.array([i.split('-')[0]+i.split('-')[1]+i.split('-')[2] for i in exchangerate.index])])
exrate.index = np.array([pd.to_datetime(str(i), format='%Y%m%d') for i in exrate.index]) #convert int index to datetime index
intersect_date = np.intersect1d(exchangerate.index,exrate.index) #find intersection dates of two table
```

```python
def create_affair_table(time,acurrency='美元',duration= 5):
    #time = exrate.index[affair_index]
    location_tmp = np.where(exchangerate.index==time)[0][0]
    location_tmp-duration,location_tmp+duration
    table = pd.DataFrame(exchangerate.loc[:,acurrency].iloc[location_tmp-duration:location_tmp+duration] 
                         #- exchangerate.loc[time,acurrency],
                )
    #print (time,exrate.loc[time])
    if pd.DataFrame(exrate.loc[time]).shape[1] != 1:  #note there might be more than one affair one day
        #print (exrate.loc[time],pd.DataFrame(exrate.loc[time]).shape)
        affair_name = ','.join([ ''.join(i) for i in exrate.loc[time,'备注'] ])
    else:
        affair_name = exrate.loc[time,'备注']
    table['affair'] = np.repeat(affair_name,duration*2).reshape(-1,1)
    table['offerdate'] = time
    return table
```

```python
def create_one_merge_table(currencyname = '美元'):
    merge_table = create_affair_table(intersect_date[0],acurrency=currencyname,duration=5)
    for i in tqdm(range(intersect_date.shape[0])):
        merge_table=pd.concat((merge_table,create_affair_table(intersect_date[i],acurrency=currencyname)),axis=0)
    merge_table['currency'] = currencyname
    merge_table.columns=['rate','affair','offerdate','currency']
    return merge_table
```

```python
exchangerate.columns
```

```python
merge_table_all = create_one_merge_table(currencyname = '美元')
for i in tqdm(exchangerate.columns[1:]):
    merge_table_all=pd.concat((merge_table_all,create_one_merge_table(currencyname = i)),axis=0)
```

```python
plot_df = pd.DataFrame(columns=['value','currency','classification','level','notes','date'])
plot_df['currency'] = np.array(np.repeat(merged_table.columns[3:],merged_table.shape[0]))
plot_df.loc[:,['classification','level','notes']] = np.array(pd.concat([merged_table]*(merged_table.shape[1]-3)).iloc[:,:3])
plot_df.loc[:,'date'] = np.array(pd.concat([merged_table]*(merged_table.shape[1]-3)).index)
for j in tqdm(range(merged_table.shape[1]-3)):
    #plot_df.loc[j*merged_table.shape[0]:(j+1)*merged_table.shape[0],'currency'] = np.repeat(merged_table.columns[j+3],merged_table.shape[0])
    for i in range(merged_table.shape[0]):
        plot_df.loc[j*merged_table.shape[0]+i,'value'] = merged_table.iloc[i,j+3]

plot_df['days'] = plot_df['date'] - plot_df['date'][0]
plot_df['intdays'] = plot_df['days'].dt.days#.astype('str')
plot_df.value = pd.to_numeric(plot_df.value)
```
