```python
#single stock
dai_bet = 6
duo_bet = 4
stock_ind = '300750'
stock_range = [-10,10]
money_init = 30

mean = (dai_bet+duo_bet)/2
standard_deviation_1 = 1/(stock_range[1]/(mean - stock_range[0]))
standard_deviation_2 = 1/(stock_range[1]/(stock_range[1] - mean))

x_values1 = np.arange(stock_range[0], mean, 0.1)
x_values2 = np.arange(mean, stock_range[1], 0.1)
y_values1 = scipy.stats.norm(mean, standard_deviation_1)
y_values2 = scipy.stats.norm(mean, standard_deviation_2)
ratio = (y_values1.pdf(x_values1)[-1])/(y_values2.pdf(x_values2)[0])
money1 = (1-y_values1.pdf(x_values1))*money_init
money2 = (1-ratio*y_values2.pdf(x_values2))*money_init

fig,ax=plt.subplots(figsize=(10,5))
ax.plot(x_values1, money1)
ax.plot(x_values2, money2)
ax.set_title('Duo: {}%, Dai: {}%, Money: {}, data: {}, stock: {}'.format(duo_bet,dai_bet,money_init,date,stock_ind))

duo_bet1,dai_bet1,money_init1,date1,stock_ind1 = duo_bet,dai_bet,money_init,date,stock_ind
```

```python
# SH index
dai_bet = -0.2
duo_bet = 0.1
stock_ind = '000001'
stock_range = [-10,10]
money_init = 30
#max_bet = max(dai_bet,duo_bet)
#min_bet = min(dai_bet,duo_bet)

mean = (dai_bet+duo_bet)/2
standard_deviation_1 = 1/(stock_range[1]/(mean - stock_range[0]))
standard_deviation_2 = 1/(stock_range[1]/(stock_range[1] - mean))

x_values1 = np.arange(stock_range[0], mean, 0.1)
x_values2 = np.arange(mean, stock_range[1], 0.1)
y_values1 = scipy.stats.norm(mean, standard_deviation_1)
y_values2 = scipy.stats.norm(mean, standard_deviation_2)
ratio = (y_values1.pdf(x_values1)[-1])/(y_values2.pdf(x_values2)[0])
money1 = (1-y_values1.pdf(x_values1))*money_init
money2 = (1-ratio*y_values2.pdf(x_values2))*money_init

fig,ax=plt.subplots(figsize=(10,5))
ax.plot(x_values1, money1)
ax.plot(x_values2, money2)
ax.set_title('Duo: {}%, Dai: {}%, Money: {}, data: {}, stock: {}'.format(duo_bet,dai_bet,money_init,date,stock_ind))
fig.savefig('stockbet/{}/betfig_SH.png'.format(date))
f= open('stockbet/{}/bet.txt'.format(date),"r+")
f.write( 'Duo: {}%, Dai: {}%, Money: {}, data: {}, stock: {}'.format(duo_bet,dai_bet,money_init,date,stock_ind))
f.write('\n')
f.write( 'Duo: {}%, Dai: {}%, Money: {}, data: {}, stock: {}'.format(duo_bet1,dai_bet1,money_init1,date1,stock_ind1))
f.close()
```
