```python
files = os.listdir('.')
```

```python
for filename in files:
    with open(filename,errors='replace') as f:
        lines = f.readlines()
        try:
            index = np.where(np.array([t[:2] for t in lines])=='to')[0][0]
            #print (lines[index])
            if lines[index][:4] =='top:':
                #print (lines[index])
                lines[index] = '#'+lines[index]
                #print (lines[index])
        except:
            print (filename +' has no top')
    with open('/Users/james/Desktop/_posts/'+filename,'w') as new_file:
        for line in lines:
            new_file.write(line)
```
