### 2021

```python
# 计算期权品种总成交量和日均成交量
df = pd.read_excel('ALLOPTIONS2021.xls',header=1)
df = df.rename(columns={'成交量(手)': '成交量'})
df['品种'] = df['合约代码'].str[0:2]
df['交易日期']=pd.to_datetime(df['交易日期'])
```

```python
def convert(item):
    if isinstance(item,str):
        if ',' not in item:
            return float(item)
        s = ''
        tmp = item.strip().split(',')
        for i in range(len(tmp)):
            s += tmp[i]
        return float(s)
```

```python
# 表格预处理，数据格式转换
df = df.fillna(0)
df['成交量'] = df['成交量'].map(convert)
df['持仓量'] = df['持仓量'].map(convert)
df1 = df[(df['交易日期']<pd.Timestamp('2021-07-01')) & (df['交易日期']>=pd.Timestamp('2021-01-01')) ]
```

```python
####计算期权品种日均成交持仓和总的成交持仓

df2 = pd.pivot_table(df1,index=['交易日期','品种'],values = ['成交量','持仓量'],aggfunc = np.sum)
df2 = df2.reset_index()
df3 = df2.groupby(['品种']).sum()
df4 = pd.pivot_table(df2,index=['品种'],values = ['成交量','持仓量'],aggfunc = np.average)
df4 = pd.merge(df4,df3, on = '品种')
```

```python
df4['成交量_x'] = df4['成交量_x'].map(lambda x:x/10000)
df4['持仓量_x'] = df4['持仓量_x'].map(lambda x:x/10000)
df4['成交量_y'] = df4['成交量_y'].map(lambda x:x/10000)
df4['持仓量_y'] = df4['持仓量_y'].map(lambda x:x/10000)
```

```python
def get_largest_volume(x, key = '成交量'):
    df = x.sort_values(by = key,ascending=True)
    return df.iloc[-1,:]
```

```python
def get_seclargest_volume(x, key = '成交量'):
    df = x.sort_values(by = key,ascending=True)
    return df.iloc[-2,:]
```

```python
##删去无用列
Op_result1 = df1.drop(['今收盘','今开盘','最高价','最低价','涨跌2','涨跌1','增减量','DELTA','行权量'],axis=1)
```

```python
##主力选取，main为保留原主力合约,select去重后用作wind导数列表
Op_process = Op_result1.groupby(['交易日期','合约月份']).sum()
Op_process = Op_process.reset_index()
Op_process['品种'] = Op_process['合约月份'].apply(lambda x:x[:2]).tolist()
```

```python
def top_1(df,columns='成交量'):
    return df.sort_values(by=columns,ascending=False)[:1]
```

```python
Op_process = Op_process.groupby(['交易日期','品种']).apply(top_1)
Op_main=Op_process.loc[:,['交易日期','品种','合约月份']]
Op_select=Op_process.loc[:,['交易日期','品种','合约月份']]
Op_select.drop_duplicates(subset='合约月份',keep='first',inplace=True)
```

```python
from WindPy import *

a="2018-01-01"
b="2021-07-15"
```

```python
##根据主力合约获取标的结算价数据
assetList=Op_select['合约月份'].apply(lambda x:x+'.czc')
assetList=assetList.tolist()
w.start()
df=w.wsd(assetList,"settle",a,b,"")
dates=pd.to_datetime(df.Times)
df=pd.DataFrame(df.Data,index=assetList,columns=dates.strftime("%Y-%m-%d"))
df1=df.stack()
df1=df1.reset_index()
df1.columns=['代码','日期','标的结算价']
df1['合约']=df1['代码'].apply(lambda x:x[:5]).tolist()
del df1['代码']
dd=w.wsd(assetList,"pre_settle",a,b,"")
dates=pd.to_datetime(dd.Times)
dd=pd.DataFrame(dd.Data,index=assetList,columns=dates.strftime("%Y-%m-%d"))
dd1=dd.stack()
dd1=dd1.reset_index()
dd1.columns=['代码','日期','标的前结算']
dd1['合约']=dd1['代码'].apply(lambda x:x[:5]).tolist()
del dd1['代码']
```

```python

##制定主力合约及标的结算价对照表
Op_main['交易日期']=pd.to_datetime(Op_main['交易日期'])
df1['日期']=pd.to_datetime(df1['日期'])
Op_main['交易日期'] = Op_main['交易日期'].apply(lambda x:x.date())
df1['日期'] = df1['日期'].apply(lambda x:x.date())

Op_1=pd.merge(Op_main,df1,left_on=['交易日期','合约月份'],right_on=['日期','合约'],how='outer')
```

```python
Op_1 = Op_1.drop(['交易日期','合约月份'],axis=1)
```

```python

##将主力合约及价格添加进原数据表格
Op_result1['交易日期']=pd.to_datetime(Op_result1['交易日期'])
Op_1['日期']=pd.to_datetime(Op_1['日期'])
#Op_result1['日期'] = Op_result1['日期'].apply(lambda x:x.date())
Op_result = pd.merge(Op_result1,Op_1,left_on=['交易日期','品种'],right_on=['日期','品种'],how='left')
##添加涨跌价差判断条件等
Op_result['涨跌']=Op_result['合约代码'].str[5:6]
Op_result['行权价格']=Op_result['合约代码'].str[6:]
Op_result['行权价格']=pd.DataFrame(Op_result['行权价格'],dtype=np.float)
Op_result=Op_result[Op_result['合约']==Op_result['合约']]
Op_result['绝对价格差']=abs(Op_result['标的结算价']-Op_result['行权价格'])
Op_result['价格差']=Op_result['标的结算价']-Op_result['行权价格']
```

```python
#col = list(qq_z_汇总.columns)
#col.remove('绝对平值')
#col = ['绝对平值']+col
#qq_z_汇总 = qq_z_汇总[col]
##选择绝对价格差最小的行
def top_2(df,columns='绝对价格差'):
    return df.sort_values(by=columns,ascending=True)[:1]
Op_result = Op_result.groupby(['日期','品种','涨跌']).apply(top_2)
##选择价格差最小的行
def top_3(df,columns='价格差'):
    return df.sort_values(by=columns,ascending=True)[:1]
Op_result = Op_result.groupby(['日期','品种','涨跌']).apply(top_3)
Op_result=Op_result.reset_index(drop=True)
dd1['日期'] = pd.to_datetime(dd1['日期'])
Op_result=pd.merge(Op_result,dd1,left_on=['日期','合约'],right_on=['日期','合约'],how='left')
```

```python

##构建虚实判断函数
def function(a,b):
    if 'C' in a and b <0:
        return '虚值'
    else:
        if 'C' in a and b>0:
            return '实值'
        else:
            if 'C' in a and b == 0:
                return '平值'
            else:
                if 'P' in a and b>0:
                    return '虚值'
                else:
                    if 'P' in a and b<0:
                        return '实值'
                    else:
                        if 'P' in a and b ==0:
                            return '平值'
                        else:
                            return '错误'
```

```python
Op_result['虚实']=Op_result.apply(lambda x: function(x.涨跌,x.价格差),axis=1)
```
